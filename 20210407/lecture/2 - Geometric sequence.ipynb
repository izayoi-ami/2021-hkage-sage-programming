### A geometric sequence is such that its next term is always generated by multiplying a fixed number K to its previous term

```python
[5 , 50, 500, 5000]  # We multiply 10 each time
```

```python
[ 3*  2^k for k in [1..10]] # We multiply 2 each time
```

# Sum of geometric sequence

$$ 1+ x + x^2 + x^3 +  \dots + x^n  = \frac{x^{n+1} - 1}{x-1}$$ for all numbers $x$
